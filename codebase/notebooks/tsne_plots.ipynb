```python
from core.semcor_bert_pipeline import *
from core.clustering import *
plt.rcParams["font.family"] = "Times New Roman"
```

```python
word_data = load_data('cover', 'v', 'semcor')
```

```python
from nltk.corpus import wordnet as wn
from textwrap import fill
```

```python
format_synsets = [i.split("Synset('")[1].strip("')") for i in word_data['sense_names']]
defns = [s + ": " + wn.synset(s).definition() for s in format_synsets]
labels = [fill(l, 30) for l in defns]
```

```python
plt.figure(figsize = (10, 8))
e = plot_embeddings(word_data['embeddings'], word_data['sense_indices'],
                labels, word_data['lemma'], shapes = ['o', 'x', '+', '^', 's'])
```

```python
color_dict, label_dict = create_dendrogram_colors(format_synsets)
plt.figure(figsize = (12, 8))
plot_dendrogram(word_data, color_dict, label_dict)
plt.ylim(0, 0.4)
plt.yticks([])
```
